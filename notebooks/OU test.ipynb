```python
%load_ext autoreload
%autoreload 2
import sys
if ".." not in sys.path:
    sys.path.append("..")

import numpy as np
import matplotlib.pyplot as plt

from simulation.diffusion import Diffusion

plt.rcParams["figure.figsize"]        = ((1 + np.sqrt(5)) / 2 * 6, 6)
plt.rcParams["figure.autolayout"]     = True
plt.rcParams["patch.force_edgecolor"] = False
plt.rcParams["axes.grid"]             = True
```

```python
t_grid = np.linspace(0, 3, 100)
size = 3
dim = 3
rng = np.random.default_rng(seed=42)

diffusion = Diffusion(t_grid=t_grid, size=size, dim=dim, rng=rng)
```

```python
W = diffusion.brownian_motion(dims=(1, 2))
OU = diffusion.ornstein_uhlenbeck(lam=[1, 2], theta=0, sigma=[1, 1], dims=(1, 2), correlation=[[1, 0.99], [0.99, 1]])
```

```python
cols = ["r", "g", "b"]

for i in range(size):
    plt.plot(t_grid, W[i, 0, :], c=cols[i])
    plt.plot(t_grid, OU[i, 0, :], "--", c=cols[i])
    # plt.plot(t_grid, OU[i, 1, :], ":", c=cols[i])
```
